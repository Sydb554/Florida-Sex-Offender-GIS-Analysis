### 2a). Converting Cleaned Florida Sex Offenders Registry Data Addresses to Coordinates: 

```python
import pandas as pd
from geopy.geocoders import GoogleV3
from geopy.exc import GeocoderTimedOut
import time

# google Maps API initialization (API code left out intentionally)
API_KEY = 'API code goes here'
geolocator = GoogleV3(api_key=API_KEY)

# file paths for input and output data
input_cleaned_file = 'FloridaSOR_Clean.csv'
output_geo_file = 'FloridaSOR_Coordinates_NEW.csv'

# function to create a full address from parts
def create_full_address(address, city, county, zip_code):
    if pd.isna(address) or pd.isna(city) or pd.isna(county) or zip_code == '00000-0000' or "Out of State" in address:
        return 'None'
    parts = [address, city, county, 'Florida', str(zip_code)]
    return ', '.join([part for part in parts if pd.notna(part)])

# function for geocoding addresses using GoogleV3 API
def geocode_address(address):
    if pd.isna(address) or address == 'None':  # check for 'None' or NaN values
        return None, None
    try:
        location = geolocator.geocode(address, timeout=10)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None

# function to process geocoding for addresses and update dataframe
def geocode_data(file_path):
    df = pd.read_csv(file_path)
    # initialize new columns for latitude and longitude
    df['Perm_Latitude'], df['Perm_Longitude'] = None, None
    df['Temp_Latitude'], df['Temp_Longitude'] = None, None
    df['Trans_Latitude'], df['Trans_Longitude'] = None, None

    # loop through each address column and geocode
    for index, row in df.iterrows():
        # geocode Permanent Address
        perm_address = row['Full_Perm_Address']
        perm_latitude, perm_longitude = geocode_address(perm_address)
        df.at[index, 'Perm_Latitude'] = perm_latitude
        df.at[index, 'Perm_Longitude'] = perm_longitude
        
        # geocode Temporary Address
        temp_address = row['Full_Temp_Address']
        temp_latitude, temp_longitude = geocode_address(temp_address)
        df.at[index, 'Temp_Latitude'] = temp_latitude
        df.at[index, 'Temp_Longitude'] = temp_longitude

        # geocode Transient Address
        trans_address = row['Full_Trans_Address']
        trans_latitude, trans_longitude = geocode_address(trans_address)
        df.at[index, 'Trans_Latitude'] = trans_latitude
        df.at[index, 'Trans_Longitude'] = trans_longitude

        # print progress for each row processed
        print(f"Processed Row {index + 1}")

        # delay because of rate limiting issues with the API
        time.sleep(1)  

    df.to_csv(output_geo_file, index=False)
    print(f"Geocoded data saved to {output_geo_file}.")

    return df

# create full addresses in the cleaned data
df_clean = pd.read_csv(input_cleaned_file)
df_clean['Full_Perm_Address'] = df_clean.apply(lambda row: create_full_address(row['PERM_ADDRESS'], row['PERM_CITY'], row['PERM_COUNTY'], row['PERM_ZIP']), axis=1)
df_clean['Full_Temp_Address'] = df_clean.apply(lambda row: create_full_address(row['TEMP_ADDRESS'], row['TEMP_CITY'], row['TEMP_COUNTY'], row['TEMP_ZIP']), axis=1)
df_clean['Full_Trans_Address'] = df_clean.apply(lambda row: create_full_address(row['TRANS_ADDRESS'], row['TRANS_CITY'], row['TRANS_COUNTY'], row['TRANS_ZIP']), axis=1)

# geocode the cleaned data and save it
geocode_data(input_cleaned_file)
```

### 2b).  Converting Cleaned Florida Public Schools Data Addresses to Coordinates: 

```python
import pandas as pd
from geopy.geocoders import GoogleV3
from geopy.exc import GeocoderTimedOut
import time

# google Maps API initialization (API code left out intentionally)
API_KEY = 'API code goes here'
geolocator = GoogleV3(api_key=API_KEY)

# function to handle geocoding
def geocode_address(address):
    try:
        location = geolocator.geocode(address, timeout=10)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None

# load CSV file
df = pd.read_csv('FloridaPublicSchools_CLEAN.csv')

# initialize columns for latitude and longitude in the DataFrame
df['Latitude'] = None
df['Longitude'] = None

# loop through each address, geocode, and update the DataFrame
for index, row in df.iterrows():
    address = row['Address']
    latitude, longitude = geocode_address(address)
    df.at[index, 'Latitude'] = latitude
    df.at[index, 'Longitude'] = longitude
    print(f"Address: {address}\nLatitude: {latitude}\nLongitude: {longitude}\n")
    time.sleep(1)  # optional: add delay to avoid rate limiting

# save the updated DataFrame to a new CSV file
output_file = 'FloridaPublicSchoolsCoordinates.csv'
df.to_csv(output_file, index=False)

print(f"Geocoded data saved to {output_file}.")
```
